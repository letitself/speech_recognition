```python
MODEL_ID = "jonatasgrosman/wav2vec2-large-xlsr-53-russian"
tokenizer = Wav2Vec2Tokenizer.from_pretrained(MODEL_ID)
model = Wav2Vec2ForCTC.from_pretrained(MODEL_ID)
```

```python
def transcriptor(some_audio): 
  audio, rate = librosa.load(some_audio, sr = 16000)
  input_values = tokenizer(audio, return_tensors = "pt").input_values
  # Storing logits (non-normalized prediction values)
  logits = model(input_values).logits
  # Storing predicted ids  
  prediction = torch.argmax(logits, dim = -1)
  # Passing the prediction to the tokenzer decode to get the transcription
  transcription = tokenizer.batch_decode(prediction)[0]
  # Printing the transcription
  return (transcription)
```

```python
from tqdm import tqdm

result_machine = {} #dict with title of wav file and model transcription

for l in tqdm(audios):
  res = transcriptor(l)
  result_machine.update({l: res})

result_machine
```
